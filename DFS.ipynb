```python
def dfs(plot, initiate, visited=None):
    if visited is None:
        visited = set()
    visited.add(initiate)

    print(initiate)

    for next in plot[initiate] - visited:
        dfs(plot, next, visited)
    return visited


plot = {'0': set(['1', '2']),'1': set(['0', '3', '4']),'2': set(['0']),'3': set(['1']),'4': set(['2', '3'])}

dfs(plot, '0')
import matplotlib.pyplot as plt

data = {'inputs' : [1020, 1300, 1500, 1800], 'time' : [310, 600, 1220, 1800]}

plt.figure(1)
plt.plot(data['time'], data['inputs'])
plt.title('time vs input length')
plt.show()
```
